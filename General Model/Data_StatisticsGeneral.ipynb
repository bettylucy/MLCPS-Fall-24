```python
import pandas as pd

oneper = pd.read_csv('unified_anomaly_dataset_1percent.csv', header=0)
fiveper = pd.read_csv('unified_anomaly_dataset_5percent.csv', header=0)
tenper = pd.read_csv('unified_anomaly_dataset_10percent.csv', header=0)
twenper = pd.read_csv('unified_anomaly_dataset_20percent.csv', header=0)

oneper['anomaly'] = oneper['anomaly'].astype(str)
fiveper['anomaly'] = fiveper['anomaly'].astype(str)
tenper['anomaly'] = tenper['anomaly'].astype(str)
twenper['anomaly'] = twenper['anomaly'].astype(str)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(oneper['anomaly'], kde=False, color='blue', label='1%', stat="frequency", fill=False, linewidth=1)
sns.histplot(fiveper['anomaly'], kde=False, color='red', label='5%', stat="frequency", fill=False, linewidth=1)
sns.histplot(tenper['anomaly'], kde=False, color='green', label='10', stat="frequency", fill=False, linewidth=1)
sns.histplot(twenper['anomaly'], kde=False, color='maroon', label='20', stat="frequency", fill=False, linewidth=1)

# Add labels and title
plt.title('Anomaly Distribution for General Case')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```
